### Feature Engineering with date

```python
### Old Format
non_top_house['date']
```

```python
non_top_house['date'] = pd.to_datetime(non_top_house['date'])
```

```python
non_top_house['date']
```

```python
non_top_house['month'] = non_top_house['date'].apply(lambda non_top_house:non_top_house.month)
```

```python
non_top_house['year'] = non_top_house['date'].apply(lambda date:date.year)
```

### Why are we not taking non_top_house['Price'].values() here

```python
X = non_top_house.drop('price',axis=1)
y = non_top_house['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```python
non_top_house.info()
```

```python
## 6480 is 30% of total entries of non_top_house
X_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation
from tensorflow.keras.optimizers import Adam
```

```python
model=Sequential()
```

```python
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(1))


model.compile(optimizer='adam',loss='mse')
```

### Why .values for y_train only

```python
model.fit(X_train,y_train.values,
          validation_data=(X_test,y_test.values)
          ,batch_size=128,epochs=400)
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
plt.figure(figsize=(12,12))
losses.plot()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,explained_variance_score
```
